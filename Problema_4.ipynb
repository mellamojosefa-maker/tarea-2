```python
# Problema 4: Calculador de perfil musical
# Evalúa al usuario según edad y gustos

edad = int(input("Ingresa tu edad: "))
musica = input("¿Te gusta la música? (si/no): ").lower()
bailar = input("¿Te gusta bailar? (si/no): ").lower()

if edad >= 18 and musica == "si" and bailar == "si":
    print("Eres una estrella de la fiesta")
elif musica == "si" and bailar == "no":
    print("Te gusta la música, pero no tanto bailar")
else:
    print("Tienes tu propio estilo")
```
