```python
# Problema 2: Números especiales del robot
# Se imprimen números del 1 al 100 divisibles por 5
# y se agrega un mensaje según condiciones

for num in range(1, 101):
    if num % 5 == 0:
        if num % 10 == 0:
            print(num, "Bip Bip")
        else:
            print(num, "Bip")
```
