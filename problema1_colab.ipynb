```python
# Problema 1: Clasificador de temperatura
# Este programa recorre temperaturas desde -30 hasta 50
# y clasifica cada valor según su condición

for temp in range(-30, 51):
    if temp < 0:
        print(temp, "Muy frío")
    elif temp < 25:
        print(temp, "Templado")
    else:
        print(temp, "Caluroso")
```
